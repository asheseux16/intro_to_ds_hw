```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

## Swiss Banknote Dataset

Six variables measured on 100 genuine and 100 counterfeit old Swiss 1000-franc
bank notes. The data stem from Flury and Riedwyl (1988). The columns correspond
to the following one label and six variables:

0. Conterfeit - 0: genuine bank notes、1: genuine bank notes (Labels)


1. Length - Length of the bank note


2. Left - Height of the bank note, measured on the left


3. Right - Height of the bank note, measured on the right


4. Bottom - Distance of inner frame to the lower border


5. Top - Distance of inner frame to the upper border


6. Diagonal - Length of the diagonal



More details: [Kaggle | Swiss banknote conterfeit detection](https://www.kaggle.com/chrizzles/swiss-banknote-conterfeit-detection)

```python
# Use [Pandas] to import data | banknotes.csv
# Put your code here 

bank = pd.read_csv("banknotes.csv")
label = bank.conterfeit

bank.head()
```

```python
# Use [Numpy function] to stack the following columns
# The 5-th column "Bottom" as X, the 6-th column "Top" as Y
# Put your code here

data_xy = np.column_stack((bank["Bottom"].values, bank["Top"].values))
data_xy
```

```python
# Use [Numpy function] to find where label = 0 / label = 1
# Put your code here

index_0 = np.where(label == 0)                                                                 # find the index where label == 0
index_1 = np.where(label == 1)                                                                 # find the index where label == 1


# Use data_xy to plot the scatter plot 
# Label = 0 → color = green | Label = 1 → color = red
# Fill X and Y in the plt.scatter 
# Remember to put on grid and legend and equal axis
# Put your code here

fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(111)
ax.set_aspect("equal")                                                                         # set axes equal 

plt.scatter(data_xy[index_0][:, 0], data_xy[index_0][:, 1], s=20, alpha=0.8, color = "green")  # plot the scatter plot
plt.scatter(data_xy[index_1][:, 0], data_xy[index_1][:, 1], s=20, alpha=0.8, color = "red")    # plot the scatter plot

plt.legend(labels = ["real banknote", "fake banknote"])                                        # add legend
plt.grid()                                                                                     # add grid
```

```python
# Use [Numpy function] to calculate the squared distance matrix between each points
# Put your code here

dist_sq = np.sum((data_xy[:, np.newaxis, :] - data_xy[np.newaxis, :, :]) ** 2, axis = -1)
```

```python
dist_sq
```

```python
def nearest_partition(dist_sq, K):
    
    # Use [Numpy function] to return the "K Nearest Neighbor"
    # Put your code here
    
    nearest = []
    
    for i in range(len(dist_sq)):                             # find the index of the K nearest neighbors
        nearest.append(dist_sq[i].argsort()[:K])

    return nearest
```

```python
# Use [Numpy function] to find where label = 0 / label = 1
# Put your code here

index_0 = np.where(label == 0)
index_1 = np.where(label == 1)


# Use data_xy to plot the scatter plot 
# Label = 0 → color = green | Label = 1 → color = red
# Fill X and Y in the plt.scatter
# Remember to put on grid and legend and equal axis
# Put your code here

fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(111)
ax.set_aspect("equal")

plt.scatter(data_xy[index_0][:, 0], data_xy[index_0][:, 1], s=20, alpha=0.8, color = "green")
plt.scatter(data_xy[index_1][:, 0], data_xy[index_1][:, 1], s=20, alpha=0.8, color = "red")

plt.legend(labels = ["real banknote", "fake banknote"])
plt.grid()
        
# Draw lines from each point to its three nearest neighbors (set K=3)
# Use some zip magic to make it happen (Hint is at below cell)
# You might need loops to generate the plot below
# set parameter [color='black', linewidth=1, alpha=0.5] when you draw lines 
# Put your code here


K = 3
nearest = nearest_partition(dist_sq, K)

for i in range(len(nearest)):                                                           # draw lines between the 3 nearest dots
    for j in range(len(nearest[i])):
        for s in range(len(nearest[i])):
            plt.plot(*zip(data_xy[nearest[i][j]], data_xy[nearest[i][s]]), color = "black", linewidth = 1, alpha = 0.5)
```

```python
# Set seed = 10 | Random select three dots from data_xy
# Save dots index in varible "target" 
# Don't need to change below three lines

random.seed(10)
target = np.array([random.randint(0, len(data_xy)) for i in range(3)])
text = ['A', 'B', 'C']


# Use [Numpy function] to find where label = 0 / label = 1
# Put your code here

index_0 = np.where(label == 0)
index_1 = np.where(label == 1)


# Use data_xy to plot the scatter plot 
# Label = 0 → color = green | Label = 1 → color = red | Target → color = blue
# Fill X and Y in the plt.scatter
# Remember to put on grid and legend and equal axis
# Put your code here

plt.figure(figsize = (12, 12))
plt.scatter(data_xy[index_0][:, 0], data_xy[index_0][:, 1], s=20, alpha=0.8, color = "green")
plt.scatter(data_xy[index_1][:, 0], data_xy[index_1][:, 1], s=20, alpha=0.8, color = "red")

plt.scatter(data_xy[target][:, 0], data_xy[target][:, 1], s=60, color = "blue") # Target dots

plt.legend(labels = ["real banknote", "fake banknote", "target dots"])
plt.grid()

# Draw lines from the selected three point to its three nearest neighbors (set K=3)
# Use some zip magic to make it happen (Hint is at below cell)
# You might need loops to generate the plot below
# set parameter [color='black', linewidth=1, alpha=0.5] when you draw lines 
# Put text A, B, C on each dots (location: Right + 0.1、 Up + 0.1)
# Put your code here


K = 3
nearest = nearest_partition(dist_sq, K)

for i in range(len(target)):                                             # draw lines between the 3 nearest dots of the targets
    for q in range(len(nearest[target[i]])):
        for s in range(len(nearest[target[i]])):
            plt.plot(*zip(data_xy[nearest[target[i]][q]], data_xy[nearest[target[i]][s]]),
                     color = "black", linewidth = 1, alpha = 0.5)


anno = ["A", "B", "C"]

for i, j in enumerate(anno):                                             # add annotations to the targets
    plt.annotate(j, (data_xy[target[i]][0] + 0.1, data_xy[target[i]][1] + 0.1),
                 weight = "bold", fontsize = 15)
plt.show()
```
