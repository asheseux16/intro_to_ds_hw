```python
import pandas as pd
import numpy as np
import math

UNCLASSIFIED = False
NOISE = -1
```

```python
def _dist(p,q):
    # compute and return the euclidean distance
    # put your code here
    
    return math.dist(p, q)
```

```python
def _eps_neighborhood(p,q,eps):
    # check if the distance between p and q is below eps
    # return True or False
    # put your code here
    
    if _dist(p, q) <= eps:
        return True
    else:
        return False
```

```python
def _region_query(m, point_id, eps):
    n_points = m.shape[1]
    seeds = []
    # find and return all points that belong to eps-neighborhood of point_id
    # put your code here
    
    for i in range(n_points):
        if _eps_neighborhood(m[:, [point_id]], m[:, [i]], eps) == True:
            seeds.append(i)
    
    return seeds
```

```python
def _expand_cluster(m, classifications, point_id, cluster_id, eps, min_points):
    # expand the cluster of cluster_id from point_id
    # identify all points belonging to cluster_id
    # update the clustering results in "classification" by assigning cluster_id to each point
    # return True if point_id is a core, False if point_id is not a core
    # write your code below
    
    if len(_region_query(m, point_id, eps)) >= min_points:
        for i in range(len(_region_query(m, point_id, eps))):
        
            if classifications[_region_query(m, point_id, eps)[i]] == False:
                classifications[_region_query(m, point_id, eps)[i]] = cluster_id
            
                _expand_cluster(m, classifications, _region_query(m, point_id, eps)[i], cluster_id, eps, min_points)
        
        classifications = pd.Series(classifications)
        classifications = classifications.drop_duplicates()
        classifications = classifications.tolist()

        return True
    else:
        return False
```

```python
def dbscan(m, eps, min_points):
    """Implementation of DBSCAN
    You can refer to wikipedia for detailed algorithm: https://en.wikipedia.org/wiki/DBSCAN
    Use Euclidean Distance as the measure
    
    Inputs:
    m - A matrix whose columns are feature vectors
    eps - Maximum distance two points can be to be regionally related
    min_points - The minimum number of points to make a cluster
    
    Outputs:
    An array with either a cluster id number or dbscan.NOISE (None) for each column vector in m
    """
    cluster_id = 1
    n_points = m.shape[1]
    classifications = [UNCLASSIFIED] * n_points
    # the main dbscan algorithm
    # put your code here
    
    for i in range(n_points):
        
        if i == 0:                                                                            #第零圈直接跑_expand_cluster
            _expand_cluster(m, classifications, i, cluster_id, eps, min_points)
        
        elif classifications[i] == False:                                                     #第一圈開始要先確認資料不屬於任何聚類
            cluster_id += 1                                                                   #建立新的聚類編號
            
            if _expand_cluster(m, classifications, i, cluster_id, eps, min_points) == True:   #確認資料可以向外擴張
                _expand_cluster(m, classifications, i, cluster_id, eps, min_points)           #開始擴張
            
            else:                                                                             #如果不能擴張
                cluster_id -= 1                                                               #就回到上一個聚類編號
                classifications[i] = -1                                                       #把分類設為outlier

    
    return classifications
```

```python
eps = 1.6
min_points = 5
a = dbscan(m, eps, min_points)
```
