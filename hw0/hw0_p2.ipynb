```python
# 2-1

import csv

with open("IMDB-Movie-Data.csv") as csvFile:            #讀入檔案
    movieData = csv.DictReader(csvFile)

    titles = []
    ratings = []
    highest = [0] * 3
    notYet = []
    

    for i in movieData:                                 #建立list儲存每個電影的title和rating
        titles.append(i["Title"])
        ratings.append(i["Rating"])
                                            
    
    for i in range(len(ratings)):                       #將string轉成float
        ratings[i] = float(ratings[i])
    
    for i in ratings:                                   #建立list儲存還沒算過的rating
        notYet.append(i)

    for i in range(3):                                  #找出最高的三個rating
        for j in notYet:
            if j >= highest[i]:
                highest[i] = j
        notYet.remove(highest[i])
    #print(highest)

        
    def ex2_1(list):                                   #由rating找出對應的電影名稱
    
        idx = []
    
        for i in range(len(list)):
            idx.append(ratings.index(list[i]))
            if len(idx) != len(set(idx)):
                idx[i] += 1
            print(titles[idx[i]])
        
    print("Top 3 movies with the highest ratings in 2016: ")
    ex2_1(highest)
```

```python
# 2-2

import csv

with open("IMDB-Movie-Data.csv") as csvFile:                      #讀入檔案
    movieData = csv.DictReader(csvFile)

    actors_prim = []
    revenue = []
    actors = []
    
    for i in movieData:                                           #建立list儲存每個電影的actors和revenue
        actors_prim.append(i["Actors"])
        revenue.append(i["Revenue (Millions)"])

    
        
    for i in range(len(actors_prim)):                             #把每一部電影的actors都建立一份清單
        #print(actors_prim[i])
        actors.append(actors_prim[i].split("|"))
    
    for i in range(len(actors)):                                  #把演員名字前後的空格去除
        for j in range(len(actors[i])):
            actors[i][j] = actors[i][j].strip()
            
    actorsNew = []
    
    for i in range(len(actors)):                                  #建立清單儲存所有的演員，重複的也只保留一次
        for j in range(len(actors[i])):
            if actors[i][j] not in actorsNew:
                actorsNew.append(actors[i][j])               

    
            
    for i in range(len(revenue)):                                 #把revenue裡的string轉成float，有缺值就略過
        if revenue[i] != "":
            revenue[i] = float(revenue[i])
        #print(revenue[i])
        
    
    
    revenueNew = [0] * len(actorsNew)
    numOfMovies = [0] * len(actorsNew)
    avgRevenue = []
    
    
    for i in range(len(actorsNew)):                              #把演員在每一部電影賺的revenue加起來存進revenueNew
        for j in range(len(actors)):
            if actorsNew[i] in actors[j]:
                if revenue[j] != "":
                    revenueNew[i] += revenue[j]
                    numOfMovies[i] += 1                          #計算演員共演了幾部電影
    
    
    
    def avg_revenue(listA, listB):                               #算出每個演員的平均收入
        for i in range(len(listA)):
            if listB[i] != 0:
                avgRevenue.append(listA[i] / listB[i])
            else:
                avgRevenue.append(-1)
          
    avg_revenue(revenueNew, numOfMovies)
    
    
    print("The actor generating the highest average revenue: ")
    print(actorsNew[avgRevenue.index(max(avgRevenue))])          #透過index函數，使用最大的平均收入找到演員
```

```python
# 2-3

import csv

with open("IMDB-Movie-Data.csv") as csvFile:           #讀入檔案
    movieData = csv.DictReader(csvFile)
    
    actors_prim = []
    ratings = []
    
    
    for i in movieData:                                #建立list儲存每個電影的actors和rating
        actors_prim.append(i["Actors"])
        ratings.append(i["Rating"])
       
    #print(actors_prim)
    
    actors = []
    
    for i in range(len(actors_prim)):                  #把每一部電影的actors都建立一份清單
        #print(actors_prim[i])
        actors.append(actors_prim[i].split("|"))
    
    for i in range(len(actors)):                       #把演員名字前後的空格去除
        for j in range(len(actors[i])):
            actors[i][j] = actors[i][j].strip()
    
    
    for i in range(len(ratings)):                      #把string轉成float
        ratings[i] = float(ratings[i])
        
    #print(actors)
    
    theMovies = []
    
    for i in range(len(actors)):                       #在演員名單中尋找 Emma Watson 的名字，找到了就將電影index傳入theMovies
        if "Emma Watson" in actors[i]:
            theMovies.append(i)
    #print(theMovies)
    
    theRatings = []
    
    for i in theMovies:                                #根據電影index找出ratings，傳入theRatings
        theRatings.append(ratings[i])
    #print(theRatings)
    
    def cal_avg(aList):                                #計算 the average rating
        
        total = 0
        
        for i in aList:
            total += i
        
        return total/len(aList)
    
    print("The average rating of Emma Watson’s movies: ", cal_avg(theRatings))
```

```python
# 2-4

import csv

with open("IMDB-Movie-Data.csv") as csvFile:                     #讀入檔案
    movieData = csv.DictReader(csvFile)
    
    directors_prim = []
    actors_prim = []
    
    
    
    for i in movieData:                                          #建立list儲存每個電影的directors和actors
        directors_prim.append(i["Director"])
        actors_prim.append(i["Actors"])
        
    
    actors = []
    
    for i in range(len(actors_prim)):                            #把每一部電影的actors都建立一份清單
        #print(actors_prim[i])
        actors.append(actors_prim[i].split("|"))
    
    for i in range(len(actors)):                                 #把演員名字前後的空格去除
        for j in range(len(actors[i])):
            actors[i][j] = actors[i][j].strip()
            
    
    directors = []
    
    for i in range(len(directors_prim)):                         #把所有導演不重複的加進directors裡
        if directors_prim[i] not in directors:
            directors.append(directors_prim[i])
            
    #print(directors)
    
    collabed = [[] for i in range(len(directors))]
    
    for i in range(len(directors)):                              #把與導演合作過的演員全部不重複的存進collabed
        for j in range(len(directors_prim)):
            if directors[i] == directors_prim[j]:
                for k in actors[j]:
                    if k not in collabed[i]:
                        collabed[i].append(k)
    
    
    
    numOfActors = [[] for i in range(len(collabed))]
    
    def find_numOfActors(aList):                                #找出跟導演合作過的演員數量
        for i in range(len(aList)):                             
            numOfActors[i].append(len(aList[i]))
    
    find_numOfActors(collabed)
    
    print("Top 3 directors who collaborate with the most actors: ")
    
    for i in range(3):
        print(directors[numOfActors.index(max(numOfActors))])
        directors.remove(directors[numOfActors.index(max(numOfActors))])
        numOfActors.remove(numOfActors[numOfActors.index(max(numOfActors))])
```

```python
# 2-5

import csv

with open("IMDB-Movie-Data.csv") as csvFile:                  #讀入檔案
    movieData = csv.DictReader(csvFile)

    
    genres_prim = []
    actors_prim = []
    
    for i in movieData:                                       #建立list儲存每個電影的genres和actors
        genres_prim.append(i["Genre"])
        actors_prim.append(i["Actors"])
    
    
    genres = []
    actors = []
    
    for i in range(len(actors_prim)):                         #把每一部電影的actors都建立一份清單
        #print(actors_prim[i])
        actors.append(actors_prim[i].split("|"))
    
    for i in range(len(actors)):                              #把演員名字前後的空格去除
        for j in range(len(actors[i])):
            actors[i][j] = actors[i][j].strip()

    for i in range(len(genres_prim)):                         #把每一部電影的genres都建立一份清單
        genres.append(genres_prim[i].split("|"))
    
    #print(genres)
    actorsNew = []
    
    for i in range(len(actors)):                              #建立清單儲存所有的演員，重複的也只保留一次
        for j in range(len(actors[i])):
            if actors[i][j] not in actorsNew:
                actorsNew.append(actors[i][j])
    
    #print(actorsNew)
    
    genresNew = []
    
    for i in range(len(genres)):                              #建立清單儲存所有的genre，重複的也只保留一次
        for j in range(len(genres[i])):
            if genres[i][j] not in genresNew:
                genresNew.append(genres[i][j])
    #print(genresNew)
    
    #print(actors)
    
    played = [[] for i in range(len(actorsNew))]
    
    for i in range(len(actorsNew)):                           #把與演員過的genre全部不重複的存進played
        for j in range(len(actors)):
            if actorsNew[i] in actors[j]:
                for k in genres[j]:
                    if k not in played[i]:
                        played[i].append(k)
    #print(played)
    
    
    numOfGenres = [[] for i in range(len(played))]
        
    for i in range(len(played)):                              #找出跟演員出演過的電影類型數量
        numOfGenres[i].append(len(played[i]))
    #print(numOfGenres)
    

    print("Top 2 actors playing in the most genres of movies: ")
    
    for i in range(2):
        print(actorsNew[numOfGenres.index(max(numOfGenres))])
        actorsNew.remove(actorsNew[numOfGenres.index(max(numOfGenres))])
        numOfGenres.remove(numOfGenres[numOfGenres.index(max(numOfGenres))])
```

```python
# 2-6

import csv

with open("IMDB-Movie-Data.csv") as csvFile:                  #讀入檔案
    movieData = csv.DictReader(csvFile)

    
    years = []
    actors_prim = []
    
    for i in movieData:                                       #建立list儲存每個電影的years和actors
        years.append(i["Year"])
        actors_prim.append(i["Actors"])
        
    #print(years)  
    #print(actors_prim)
    
    for i in range(len(years)):
        years[i] = int(years[i])
    
    actors = []
    
    for i in range(len(actors_prim)):                         #把每一部電影的actors都建立一份清單
        #print(actors_prim[i])
        actors.append(actors_prim[i].split("|"))
    
    for i in range(len(actors)):                              #把演員名字前後的空格去除
        for j in range(len(actors[i])):
            actors[i][j] = actors[i][j].strip()
            
    
    
    actorsNew = []
    
    for i in range(len(actors)):                              #建立清單儲存所有的演員，重複的也只保留一次
        for j in range(len(actors[i])):
            if actors[i][j] not in actorsNew:
                actorsNew.append(actors[i][j])               

    
    theYears = [[] for i in range(len(actorsNew))]
    
   

    for i in range(len(actorsNew)):                           #把每個演員出演過的年份儲存到theYears
        
        for m in range(len(actors)):
            
            for n in range(len(actors[m])):
                if actorsNew[i] == actors[m][n]:
                    theYears[i].append(years[m])              
                    
    #print(theYears)
    
    theGapOfYears = []
    
    def gapOfYears(aList):                                    #算出每個演員的gap of years
        for i in aList:
            theGapOfYears.append(max(i) - min(i))
    gapOfYears(theYears)
    #print(theGapOfYears)                                     #find out that the largest gap of years is 10                             
    
    

     
    print("Top 3 actors whose movies lead to the largest maximum gap of years: ") 
    
    
    actMaxGOY = []
    
    for i in range(len(theGapOfYears)):                       #把gap of years等於10的演員找出來
        if theGapOfYears[i] == 10:
            actMaxGOY.append(actorsNew[i])
        #if len(actMaxGOY) == 3:
           # break
            
    print(actMaxGOY)
    #print()
    #print("(Actually many of the actors have the same gap of years)")
```
